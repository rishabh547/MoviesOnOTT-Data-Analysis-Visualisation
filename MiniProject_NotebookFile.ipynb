# 2. Graphical visualisation of Data

```python
netflix_movies=df.loc[df.Netflix==1]
total_netflix=len(netflix_movies)

prime_movies=df.loc[df.iloc[0:,8]==1]
total_prime=len(prime_movies)

hulu_movies=df.loc[df.Hulu==1]
total_hulu=len(hulu_movies)

disney_movies=df.loc[df.iloc[0:,9]==1]
total_disney=len(disney_movies)
def get_byYear(movie_dic):
    movie_years=np.array(movie_dic.iloc[0:,2],dtype=np.float)
    movies_by_year={}
    movies_after2015=[]
    for i in range(len(movie_years)):
        if(movie_years[i]>=2015):
            movies_after2015.append(movie_years[i])
    movies_after2015.sort()
    movies_after2015=np.array(movies_after2015,dtype=np.str)
    for i in range(len(movies_after2015)):
        if movies_after2015[i] in movies_by_year:
            movies_by_year[movies_after2015[i]]+=1
        else:
            movies_by_year[movies_after2015[i]]=1
    return movies_by_year

hulu_after2015=get_byYear(hulu_movies)
netflix_after2015=get_byYear(netflix_movies)
disney_after2015=get_byYear(disney_movies)
prime_after2015=get_byYear(prime_movies)
index=np.arange(6)
height1=list(netflix_after2015.values())
height2=list(prime_after2015.values())
height3=list(disney_after2015.values())
height4=list(hulu_after2015.values())
bar_width=0.25

plt.bar(index-1.5*bar_width, height1, label="Netflix", width=bar_width,align='center')
plt.bar(index-0.5*bar_width, height2, label="Amazon Prime", width=bar_width,align='center')
plt.bar(index+0.5*bar_width, height3, label="Disney+", width=bar_width,align='center')
plt.bar(index+1.5*bar_width, height4, label="Hulu", width=bar_width,align='center')

tick_label=list(netflix_after2015.keys())
plt.xticks(index, tick_label)
plt.title("Number of Movies released after 2015")
plt.legend()
plt.show()
```

```python
def movie_genre(dic,num):
    genres=np.array(dic.iloc[0:,11],dtype=np.str)
    movie_genres={}
    for i in range(len(genres)):
        if genres[i] not in movie_genres:
            movie_genres[genres[i]]=1
        else:
            movie_genres[genres[i]]+=1
    movie_genres10={}
    for i in movie_genres.keys():
        if movie_genres[i]>=num:
            movie_genres10[i]=movie_genres[i]
    return movie_genres10

def avg_imdb(dic,genre):
    imdbs_df=dic.loc[dic.Genres==genre]
    imdbs=np.array(imdbs_df.iloc[0:,4],dtype=np.float)
    avg=np.mean(imdbs)
    return avg

def avg_of_genre_imdb(dic,num):
    pgenres=movie_genre(dic,num)
    newdic={}
    li=[]
    for i in pgenres.keys():
        if(i=="unknown"):
            continue
        a=avg_imdb(dic, i)
        li.append(pgenres[i])
        li.append(a)
        newdic[i]=li
        li=[]
    return newdic

def list_of(dic):
    size=[]
    aimdb=[]
    genre=list(dic.keys())
    genre.sort()
    for i in range(len(genre)):
        size.append(dic[genre[i]][0])
        aimdb.append(round(dic[genre[i]][1],2))
    return genre,size,aimdb

def plot_plots(movies,num):
    service_genres=avg_of_genre_imdb(movies,num)
    s_genre,s_size,net_imdb=list_of(service_genres)
    explode = ([0.05]*len(s_size))
    fig2, ax2 = plt.subplots()
    ax2.pie(s_size, explode=explode, labels=s_genre, autopct='%1.1f%%',
        shadow=True)
    ax2.axis('equal')
    plt.title("Percentage of movies of Top Genres")
    plt.show()
    plt.bar(range(len(s_genre)),net_imdb, color ='red',  
        width = 0.5)

    plt.xticks(range(len(s_genre)),s_genre,rotation=90)

    for i in range(len(s_genre)):
        plt.text(x = i-0.25, y = net_imdb[i]+0.1, s = net_imdb[i], size = 6)
  
    plt.xlabel("Most Popular Genres") 
    plt.ylabel("Average Imdb Rating") 
    plt.title("Average Imdb of Top Genres in Movies") 
    plt.show() 
    
```

```python
print("Netflix")
plot_plots(netflix_movies,35)
print()

print("Amazon Prime")
print()
plot_plots(prime_movies,150)

print("Disney+")
print()
plot_plots(disney_movies,10)

print("Hulu")
print()
plot_plots(hulu_movies, 25)
```
